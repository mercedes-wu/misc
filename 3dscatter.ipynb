```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
import pandas as pd

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = dash.Dash(__name__, external_stylesheets=external_stylesheets)

df = px.data.iris()

def generate_table(df, ):
    return html.Table([
        html.Thead(
            html.Tr([html.Th(col) for col in df.columns])
        ),
        html.Tbody([
            html.Tr([
                html.Td(df.iloc[i][col]) for col in df.columns
            ]) for i in range(min(len(df), len(df)))
        ])
    ])

def update_graph(nrows):
    temp_df = df.head(nrows)
    fig = px.scatter_3d(temp_df, x='sepal_length', y='sepal_width', z='petal_width',
              color='species')
    return fig

available_indicators = df.columns[0:4]

app.layout = html.Div(children=[
    html.H1(
        children='Dash App: 3-D Sepal Analysis',
        style={
            'textAlign': 'center',
        }
    ),

#     generate_table(pd.DataFrame()),
    
    dcc.Graph(
        id='multi-plot',
    ),
        
    dcc.Dropdown(
        id='multi-select dropdown',
        options=[{'label': i, 'value': i} for i in available_indicators],
        value=[],
        multi=True
    ),
    
    html.Div(id='my-output'),
    
])


@app.callback(
    Output('multi-plot', 'figure'),
    Input('multi-select dropdown', 'value')
)
def update_graph(input_value):
    feat_list = input_value.copy()
    feat_list.append('species')
    temp_df = df[feat_list]

    if len(feat_list)-1 == 1:
        fig = px.violin(temp_df, y=feat_list[0], x='species', color="species", box=True, points="all",
                        hover_data=temp_df.columns, template='plotly_white')
        
    elif len(feat_list)-1 == 2:
        fig = px.scatter(temp_df, x=feat_list[0], y=feat_list[1], color='species', opacity=.7,template='plotly_white')
    
    elif len(feat_list)-1 == 3:
        fig = px.scatter_3d(temp_df, x=feat_list[0], y=feat_list[1], z=feat_list[2],
                            color='species', opacity=.7, template='plotly_white')
        fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
        
    elif len(feat_list)-1 == 4:
        fig = px.scatter_3d(df, x=feat_list[0], y=feat_list[1], z=feat_list[2],
              color=feat_list[3], symbol='species', opacity=0.7, template='plotly_white')
        fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
        
    else: 
        temp_df = pd.DataFrame()
        fig = px.scatter(temp_df, template='plotly_white')
        
    return fig
        
   
@app.callback(
    Output('my-output', 'children'),
    Input('multi-select dropdown', 'value')
)
def print_value(input_value):
     return 'Output: {}'.format(input_value)
app.run_server()
```
